```python
import pandas as pd
```

```python
df =pd.read_csv(r'/Users/joblethomas/Desktop/Learning/Crime_Data_from_2020_to_Present.csv')
```

```python
df.head()
```

```python
# Dropping Unnecessary Columns 
```

```python
df.drop(['Premis Cd','Weapon Used Cd','Status','Mocodes', 'AREA', 'DR_NO', 'Rpt Dist No', 'Crm Cd', 'Part 1-2', 'Status Desc', 'Crm Cd 1', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4','LOCATION', 'Cross Street' ], inplace=True, axis=1)
```

```python
df.drop("Victim's Descent" , inplace=True, axis=1)
```

```python
# Replacing String Values 
```

```python
df["Vict Sex"].replace(['F','M','X'],['Female','Male','N/A'], inplace = True )
```

```python
df = df[df['Vict Sex'].isin(['Female', 'Male', 'N/A'])]
```

```python
# Filling in Blank Spaces
```

```python
df["Weapon Desc"].fillna('None', inplace = True )
```

```python
df["Premise Description"].fillna('Unknown', inplace = True )
```

```python
#Renaming Column Headers
```

```python
df.rename(columns={"Date Rptd": "Date Reported", "DATE OCC": "Date Occurrence", "TIME OCC": "Time Occurrence", "AREA NAME": "Area Name","Crm Cd Desc": "Crime Description", "Vict Age": "Victim's Age", "Vict Sex": "Victim's Sex", "Vict Descent": "Victim's Descent", "Premis Desc": "Premise Description", "Weapon Desc": "Weapon Description", "LAT": "Latitude", "LON": "Longitude" }, inplace=True)
```

```python
# Changing the Data Type
```

```python
df['Date Reported'] = pd.to_datetime(df['Date Reported'])
```

```python
df['Date Occurrence'] = pd.to_datetime(df['Date Occurrence'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['Time Occurrence'] = df['Time Occurrence'].astype(str).str.zfill(4)  # Ensure it's 4 digits
df['Time Occurrence'] = pd.to_datetime(df['Time Occurrence'], format='%H%M').dt.strftime('%H:%M')
df.head()
```

```python
# Removing Certain Rows of Data 
```

```python
df.drop(df[df["Victim's Age"].isin([-1, -2, -3])].index, inplace=True)
```

```python
#Saving the Cleaned Data
```

```python
df.to_csv('Los_Angeles_Crime_Analysis.csv')
```
